```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

**(1) Finding duration of majority of the movies in the 90s**

```python
# Filtered movies from 1990 to 2000 (90s decade)
movies_1990s_df = netflix_df[(netflix_df['release_year']>=1990) & (netflix_df['release_year']<=1999)]
movies_1990s_df
```

```python
#plot graph to see duration of 90s movies
plt.clf()
plt.hist(movies_1990s_df['duration'])
plt.xlabel('Duration in Minutes')
plt.ylabel('Number of Movies')
plt.title('90s Movies duration distribution')
plt.show()
```

```python
#Majority movies had average duration of 110mins(rounded)
duration = 110 
duration
```

**(2) Finding number of short action movies(<90mins) in 90s**

```python
#filtering for short & action 90s movies
short_action_df = movies_1990s_df[(movies_1990s_df['genre']=='Action')&(movies_1990s_df['duration']<90)]
short_action_df
```

```python
#getting total number of short & action 90s movies
short_movie_count = len(short_action_df)
short_movie_count
```
